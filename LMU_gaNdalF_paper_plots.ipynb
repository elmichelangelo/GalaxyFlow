```python
import pandas as pd
from Handler import *
import os
```

```python
def apply_deep_cuts(path_data, filename_master_cat, data_frame):
    """"""
    data_frame = flag_cuts(data_frame=data_frame)
    data_frame = mask_cut(
        data_frame=data_frame,
        master=f"{path_data}/{filename_master_cat}"
    )
    return data_frame
```

```python
path_data = "/project/ls-gruen/users/patrick.gebhardt/data/gaNdalF_paper_catalogs"
path_data_master_cat = "/project/ls-gruen/users/patrick.gebhardt/data/gaNdalF"
filename_clf_balrog = "2024-10-28_08-14_balrog_clf_Test_sample.pkl"
filename_clf_gandalf = "2024-10-28_08-14_gandalf_clf_Test_sample.pkl"
filename_master_cat = "Y3_mastercat_02_05_21.h5"
path_save_plots = "/home/p/P.Gebhardt/Output/gaNdalF_paper"
```

```python
df_balrog_clf = pd.read_pickle(f"{path_data}/{filename_clf_balrog}")
df_gandalf_clf = pd.read_pickle(f"{path_data}/{filename_clf_gandalf}")
```

```python
df_balrog_clf_deep_cut = apply_deep_cuts(path_data_master_cat, filename_master_cat, df_balrog_clf)
df_gandalf_clf_deep_cut = apply_deep_cuts(path_data_master_cat, filename_master_cat, df_gandalf_clf)
```
